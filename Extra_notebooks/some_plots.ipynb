```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
from sqlalchemy import create_engine
from configparser import ConfigParser
```

```python
config = ConfigParser()
config.read('database.ini')
conf = config['postgresql']

engine = create_engine(f"postgresql://{conf['username']}:{conf['password']}@{conf['hostname']}/{conf['database']}")

with engine.connect() as conn:
    df_main = pd.read_sql_table('location_ping', conn)
    df_time = pd.read_sql_table('time', conn)
```

```python
tags = list(df_main['tag_id'].drop_duplicates())
str(tags)
```

```python
df_1 = df_main[df_main['tag_id']=='AM105']
df_2 = df_main[df_main['tag_id']=='AM107']
fig = px.line_mapbox(df_1, lat="latitude", lon="longitude",  zoom=7, height=300)
fig.add_trace(px.line_mapbox(df_2,lat='latitude', lon='longitude').data[0])
fig.update_traces(line=dict(color='purple'), selector=0)
fig.update_traces(line=dict(color='red'), selector=1)
fig.update_layout(mapbox_style="stamen-terrain")
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
fig.show()
```
